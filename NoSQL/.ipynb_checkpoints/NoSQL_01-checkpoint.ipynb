RDBMS : 관계형 데이터베이스 관리 시스템(체계)
- 현존하는 가장 오래된 데이터베이스 관리 시스템
- 안정적, 가성비, 대중적
- 시대 변화 -> 문제 지적발생
- MySQL 활용 RDBMS -> NAVER 포털사이트 영화배우 이름 검색 -> 데이터 수집 -> DB 저장
- python > mysql (신체, 소속사, 가족, 수상, 사이트, 활동, 이미지, url)
- 정해진 패턴을 기반으로 데이터 수집, 저장 // 데이터 다양성 & 방대 => 빅데이터 시대
- 디비 > (패턴)스키마 > 테이블 > 데이터 (시대 x)

NoSQL : Not Only SQL (전통적인 SQL 문법만 의존하는 관계형데이터베이스 관리 시스템 탈피)
- 고정된 Scheme 포기 / JOIN 문법자체가 존재 불가능해

RDBMS : 최초에 한 번 혹은 아주 가끔씩만 데이터 저장 / 저장된 데이터를 수정하지 않고, 그냥 읽어오는 업무 프로세스 최적 (많이 사용) -> MySQL, Oracle, PostgreSQL, SQLlite

NoSQL : 자주 빈번하게, 데이터의 입출력 가능해야하는 상황 (DB -> ecommerce / 멤버십 관리 필수 / 금융권)
-> MongoDB
-> JSON 형태를 띄고 있음 => JavaScript Object Notation => JavaScript => 웹 브라우저가 인식할 있는 유일한 프로그래밍 언어 <-> Backend 영역
-> BSON 문법 체계 => 빅데이터 // 데이터 처리 속도 (0, 1)

RDBMS : DB > Table > Scheme > Row > Column > Field > PK | FK
NoSQL : DB > Collection > Document (key, value) > Field > _id (= PK)
