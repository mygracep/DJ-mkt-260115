```python
import pandas as pd

df1 = pd.DataFrame({
    "id": [1, 2, 3],
    "customer_id": [1, 2, 3],
    "customer_name": ["Robert", "Peter", "Dave"]
})

df1
```

```python
import pandas as pd

df2 = pd.DataFrame({
    "id": [1, 2, 4],
    "order_id": [100, 200, 300],
    "order_date": ["2021-01-21", "2021-02-03", "2020-10-01"]
})

df2
```

```python
# concat() : 서로 다른 두 개의 데이터 프레임을 물리적으로 합쳐주는 메서드 (디폴트는 상.하 위치로)
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis=1) # 0 -> 상하
```

```python
# merge() : 서로 다른 두 개의 데이터프레임 안에 동일한 이름을 가진 컬럼을 기준으로 합침
```

```python
pd.merge(df1, df2) # 각 데이터프레임의 필드명을 왼쪽부터 조회 -> 상호 동일한 필드명이 조회(선택)
```

```python
pd.merge(df1, df2, on="id")
```

```python
# merge(데이터프레임1, 데이터프레임2, 연결필드, how=결합방법)
# inner : SQL 이너조인과 동일한 방법으로 연결
# outer : 각 데이터 프레임의 인덱스값을 기준으로 서로 다른 데이터 프레임을 연결하겠다는 의미
# left : SQL LEFT OUTER JOIN
# right : SQL RIGHT OUTER JOIN
```

```python
import pandas as pd

pd.merge(df1, df2, on="id", how="inner") # 애초에 디폴트값이 inner join
```

```python
import pandas as pd

pd.merge(df1, df2, on="id", how="outer")
```

```python
import pandas as pd

pd.merge(df1, df2, on="id", how="left")
```

```python
import pandas as pd

pd.merge(df1, df2, on="id", how="right")
```

```python
df1 = df1.set_index("id")
```

```python
df1
```

```python
df2 = df2.set_index("id")
df2
```

```python
pd.merge(df1, df2, left_index=True, right_index=True) # 사실 how=inner 생략되어있음
```

```python
pd.merge(df1, df2, how="outer", left_index=True, right_index=True)
```
