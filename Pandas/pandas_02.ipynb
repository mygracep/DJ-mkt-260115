```python
import csv

data_file = open("00_data/USvideos.csv", "r", encoding="utf-8-sig")
data_lines = csv.reader(data_file, delimiter=",")
for data_line in data_lines :
    print(data_line)
    break
data_file.close()
```

```python
with open("00_data/USvideos.csv", "r", encoding="utf-8-sig") as reader_csv :
    reader = csv.reader(reader_csv, delimiter=",")

    for row in reader :
        print(row)
        break
```

```python
# csv 파일을 리스트 형태로 값을 삽입하는 방법
import csv

with open("00_data/tmp_csv.csv", "w", encoding="utf-8-sig") as writer_csv :
    writer = csv.writer(writer_csv, delimiter=",")
    writer.writerow(["love"] * 3 + ["banana"])
    writer.writerow(["apple", 2])
    writer.writerow(["dog", "cat", "mouse"])
```

```python
# 앞으로 다룰 행.열 기반의 테이블 안에 어떤 값을 넣을 때,
# 1개의 행 구분짓는 자료형 => 리스트
```

```python
# csv 파일을 딕셔너리 형태로 값을 삽입하는 방법
import csv

with open("00_data/tmp_csv.csv", "w", encoding="utf-8-sig") as writer_csv :
    field_name_list = ["First_Name", "Last_Name"]
    writer = csv.DictWriter(writer_csv, fieldnames=field_name_list)
    writer.writeheader()
    writer.writerow({"First_Name": "David", "Last_Name": "Park"})
    writer.writerow({"First_Name": "Dave", "Last_Name": "Lee"})
    writer.writerow({"First_Name": "Robin", "Last_Name": "Hood"})
```

```python
import csv

with open("00_data/tmp_csv.csv", "r", encoding="utf-8-sig") as reader_csv :
    reader = csv.DictReader(reader_csv)
    for row in reader :
        print(row["First_Name"], row["Last_Name"])
```
