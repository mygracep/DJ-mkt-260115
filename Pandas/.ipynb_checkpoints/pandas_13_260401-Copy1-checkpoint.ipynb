```python
# Namibia = 국가명, iso2 = NA, read_csv -> NA, N/A, #NA, na 등 -> null
country_info = pd.read_csv("COVID-19-master/csse_covid_19_data/UID_ISO_FIPS_LookUp_Table.csv", encoding="utf-8-sig", keep_default_na=False, na_values="")
country_info.head()
country_info.info()
```

```python
country_info = country_info[["iso2", "Country_Region"]]
```

```python
country_info.head()
```

```python
country_info.shape
```

```python
country_info = country_info.drop_duplicates(subset="Country_Region", keep="last")
```

```python
country_info.shape
```

```python
doc_final_country = pd.merge(df_confirmed, country_info, on="Country_Region", how="left")
```

```python
iso2 = "KR"
PATH = "https://public.flourish.studio/country-flags/svg/"
flag_link = f"{PATH}{iso2.lower()}.svg"

flag_link
```

```python
def create_flag_link(column) :
    PATH = "https://public.flourish.studio/country-flags/svg/"
    flag_link = f"{PATH}{column.lower()}.svg"
    print(flag_link)
    return flag_link
```

```python
doc_final_country["iso2"] = doc_final_country["iso2"].apply(create_flag_link)
```
