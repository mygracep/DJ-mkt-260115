```python
# 데이터가 어느 분야(산업)에서 나온 데이터인가!!!
# 의류, 식품, 주류, 홈쇼핑, 이커머스, IT, 가구, 화장품, 자동차....
# 그래서 첫 회사, 첫 사회진입 분야 중요 (분야, 산업계 -> 마케팅 직군)
# 도메인 학습

# 이커머스 => 학습
# 온라인 비즈니스의 한 종류 (고객이 온라인을 통해서 상품 조회. 구매. 결제 실행 -> 배송단계 -> 고객 상품 도달 -> 리뷰.평점 -> 재구매 발생)
# 데이터 사이언티스트, 데이터 분석가 => 채용공고 (새로운 업계, 분야를 즐겁게, 오픈마인드로 학습할 마인드, 자세)
# 온라인 비즈니스 타입
# B2C : Business To Customer (기업과 고객) : 쿠팡, SSG
# B2B : Business To Business (기업과 기업) : MRO, 폐쇄몰
# B2BC : Business To Business Customer (기업과 기업 & 고객) : BBQ, 아리따움
# D2C : Direct To Customer (기업이 직접 고객 데이터를 수집할 수 있는 비즈니스 모델)
# 특정 지역.국가 문화적 특징 : 함부로 머리를 자르지 않는다, 머리카락 귀하게 (영혼) -> 헤어케어
# 바디케어, 향수
# 우리나라 -> 빨리빨리 민족 -> 산업반영 -> 상품 구매 결정 -> 단계 최소
# 일본 -> 라쿠텐 (구매 -> 정말이요? -> 지역)
# 미국 -> 아마존 (구매 -> 쿠폰적용 -> 발급 -> 온라인, 오프라인 -> 지역매장)
# 브라질 -> olist (결제 단계, 카테고리별, 상품별 단계)
# 쇼핑몰 : 승리님 카드 -> 결제 날짜
# 판매자 : 승리님이 결제한 상품 배송상태 / 판매.배송 승인 날짜
# 배송 : 상품(기사 날짜) -> 수하물센터(상품수취) -> 수하물센터(상품배송지) -> 지역센터(세부배송지) -> 고객전달날짜


# olist : 비즈니스 중계 플랫폼 형태
# A라는 업체가 온라인으로 상품을 판매하고 싶어함 (직원, 마케팅, 등록, 배송, 물류 -> x)
# A <-> olist (계약)
# A -> olist (물류창고 상품 입고)
# A -> olist (상품 등록 메뉴얼 -> 상품등록)
# A의 상품, 주문 -> olist에서 주문확인 & 물류창고 입고 A상품 조회 -> 고객 배송
# A의 고객, 상품 수령 -> 이메일을 통해서 상품 주문.배송 관련 후기 작성 요청
# A의 고객, 후기 작성 & 등록
# A는 olist 계약당시, 서명한 정산날짜에 A의 고객이 입금한 결제금액, 수수료 제외하고 입금
```

```python
# 신사업, 스타트업, 서비스 (NSM)
# 기존 사업 히스토리 기반, 더 좋은 서비스 (Problem Definition -> NSM)
# 위에서 산정한 목표에 도달하기 위해서 우리가 알아야하는 최소한의 정보?
```
