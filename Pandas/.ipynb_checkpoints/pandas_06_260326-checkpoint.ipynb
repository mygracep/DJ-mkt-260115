```python
# 데이터 수집 -> 데이터 저장 -> 데이터 구조 -> 데이터 분석
# python > pandas
# pandas => 행.열 기반으로 되어있는 테이블형 데이터를 취급하는 전문 라이브러리 모듈
# 데이터를 꼭 테이블형 취급?
# 데이터를 취급.분석 주체 = 사람 = 익숙한 형태의 구조 = 행.열 구조 친숙
# 가로.세로 개념 => 값을 나열 => 패턴, 공통점 찾아내는데 익숙
# 엑셀, 스프레드 시트 => 행.열
# 행.열을 다루는데 있어서 굉장히 다양한 속성.메서드 등의 기능
# 기본적인 raw data(원본 데이터) => 행.열 기반으로 변환 => 정제 => 전처리 가능
# 어떤 데이터 생성 : 오타, 결측치(null), 행.열 위치 바뀌어져, 필드명 빠져있다 => 상황.수정 작업 => 정제 (cleaning)
# 정제 작업을 포함한 정상적인 데이터 사용을 위한 포괄적인 작업 => 전처리
# 전처리 (문자열 타입 -> 날짜 타입, 성별(남.여) 표기 -> 0, 1) > 정제
# EDA, M/L, D/L, pytorch, numpy
```

```python
import pandas as pd

# pandas에서 취급한 데이터 종류 : 2가지
# 시리즈 = series = 1차원 데이터
# 데이터프레임 = dataframe = 2차원 데이터 => 1차원 데이터가 여러개 묶여있는 형태
# 각 차원 데이터를 표기할 때, 유사 리스트의 형태

seriesdata = pd.Series([70, 60, 70], index=["국어", "영어", "수학"])
seriesdata

# 인덱스 = 파이썬 프로그래밍 언어, 리스트 등의 이터러블한 객체값을 조회 => 인덱스 값이 생성, SQL | NoSQL => 방대한 데이터 특정 데이터 조회찾아
```

```python
seriesdata.index # dtype = int64, object = 문자열, 2개 이상의 데이터 타입 혼합
```

```python
seriesdata.index = ["미술", "음악", "체육"]
```

```python
seriesdata
```

```python
seriesdata.values
```

```python
print(seriesdata["미술"], seriesdata.iloc[0])
```

시리즈 데이터 타입 종류
- object : str | 혼용데이터 (문자열 + 정수, 정수 + 논리, 문자열 + 논리)
- int64 : 정수 (64비트)
- float64 : 부동소수점을 가지고 있는 실수 (64비트)
- bool : True, False
- datetime64 : 날짜.시간 타입
- 20260326 => 정수, 1, 11, 2, 3, 4, 5 
- 날짜 => 시계열데이터 => 시간 경과 값이 변경되고 있다는 것을 확인할수있는 데이터 타입
