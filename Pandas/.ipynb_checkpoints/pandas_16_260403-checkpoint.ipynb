```python
import plotly.graph_objects as go
import numpy as np

data = np.random.rand(500) # 0~1 사이의 실수자료형태의 난수값을 500개 생성

fig = go.Figure()

fig.add_trace(
    go.Histogram(
        x = data,
        nbinsx = 30,
        name="히스토그램 데이터",
        showlegend=True
    )
)

fig.update_layout(
    title="히스토그램 그래프",
    xaxis_title="데이터",
    yaxis_title="빈도"
)

fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

data1 = np.random.rand(100)
data2 = np.random.rand(100)

fig = go.Figure()

fig.add_trace(
    go.Box(
        y=data1,
        name="비교열1",
        showlegend=True
    )
)

fig.add_trace(
    go.Box(
        y=data2,
        name="비교열2",
        showlegend=True
    )
)

fig.update_layout(
    title="박스플롯 그래프",
    xaxis_title="비교열 데이터"
)

fig.show()
```
