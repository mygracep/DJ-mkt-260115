```python
# EDA
# 1. 데이터 출처, 주제
# 2. 데이터 사이즈
# 3. 데이터 속성 특징 확인
# 1) 위치 & 변이추정
# 2) 수치 & 범주
# 3) 속성간 상관관계 확인
# 피어슨 상관계수
# heatmap 차트, scatter 차트
```

```python
# EDA > HEATMAP

import pandas as pd
doc = pd.read_csv("COVID-19-master/csse_covid_19_data/csse_covid_19_daily_reports/04-01-2020.csv", encoding="utf-8-sig")
```

```python
doc
```

```python
doc2 = doc.corr(numeric_only=True) # correlation (상관관계) coefficient (계수)
```

```python
import plotly.express as px

fig_px = px.imshow(
    doc2,
    text_auto=".2f", #True, False
    aspect="auto", # auto, equal
    title="Heatmap (Plotly Express)"
)

fig_px.show()
```

```python
doc2
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(
    go.Heatmap(
        x=doc2.columns,
        y=doc2.index,
        z=doc2.values,
        colorscale=[(0, "#F2B077"), (0.5, "#B0D1D9"), (1, "#41A8BF")], # Viridis, Cividis, Plasma, Inferno, Magma, Rainbow, Jet, Picnic, Portland, Electric
        colorbar={
            "title": "상관관계정도",
            "thickness": 40,
            "tickvals":[-1, 0, 1],
            "ticktext":["음의상관관계", "상관관계없음", "양의상관관계"],
            "len": 1
        },
        zmin=-1,
        zmax=1,
        text=doc2.values,
        texttemplate="%{z:.2f}",
        hovertemplate="X: %{x}<br>Y: %{y}<br>Value: %{z:.2f}<extra>Custom Info</extra>"
    )
)

fig.update_layout(title="Heatmap (Plotly Graph Objects)")

fig.show()
```

```python
# EDA > SCATTER

import pandas as pd
doc = pd.read_csv("COVID-19-master/csse_covid_19_data/csse_covid_19_daily_reports/04-01-2020.csv", encoding="utf-8-sig")
```

```python
doc.head()
```

```python
doc.corr(numeric_only=True)
```

```python
# python > pandas + plotly

# import plotly.express as px 
import plotly.io as pio

pio.templates
```

```python
import plotly.express as px

fig_px = px.scatter(
    doc,
    x="Confirmed",
    y="Active",
    title="Scatter Plot (Plotly Express)"
)

fig_px.update_layout(template="xgridoff")

fig_px.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=doc["Confirmed"],
        y=doc["Active"],
        mode="markers"
    )
)

fig.show()
```
