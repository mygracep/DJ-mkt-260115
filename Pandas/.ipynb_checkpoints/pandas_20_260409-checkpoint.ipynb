```python
# 피봇테이블 => 기존 데이터에서 추가적으로 필터링을 해서 값을 찾아오는 것

import pandas as pd

doc_covid = pd.read_csv("COVID-19-master/csse_covid_19_data/csse_covid_19_daily_reports/04-01-2020.csv", encoding="utf-8-sig")
doc_covid.head()
```

```python
import pandas as pd

doc_covid2 = pd.pivot_table(doc_covid, index=["Country_Region"], values=["Confirmed"], aggfunc={"Confirmed": "median"}, fill_value=0)

doc_covid2.head()
```

```python
import pandas as pd

doc_covid3 = pd.pivot_table(doc_covid,
               index=["Country_Region"],
               values=["Confirmed", "Deaths"],
               aggfunc={"Confirmed": "sum", "Deaths": "mean"},
               fill_value=0,
                margins=True,
                margins_name="Total"
              )

# doc_covid3.head()
```

```python
doc_covid3.style.format({
    "Confirmed": "{:,}",
    "Deaths": "{:,.2f}"
})
```

```python
doc_covid3[doc_covid3.index == "Total"]
```
