```python
import plotly.graph_objects as go

# Figure 생성
fig = go.Figure()

# Figure 안에 그래프 작성 선언
fig.add_trace(
    # 그래프 타입 결정 (* 동시에 여러 그래프를 한 번에 그릴 수 있음)
    go.Scatter(
        # 그래프 내 데이터 삽입
        x=[1, 2, 3],
        y=[1, 2, 3]
    )
)

# 플로틀리 익스프레스 처럼 추가 옵션 설정
# fig.update_layout(), fig.update_xaxes(), fig.update_yaxes()

fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

x_data = np.random.rand(100) # 0~1사이의 실수형태의 난수(랜덤)값 100개
y_data = np.random.rand(100) # 0~1사이의 실수형태의 난수(랜덤)값 100개

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x = x_data,
        y = y_data,
        mode="markers",
        marker=dict(
            size=12,
            color="orange",
            opacity=0.7
        ),
        name="산점도 범례",
        showlegend=True
    )
)

fig.update_layout(
    title="산점도 그래프",
    xaxis_title="x축",
    yaxis_title="y축" 
)

fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

x_data = np.random.rand(100)
y_data = np.random.rand(100)
categories = np.random.choice(["A", "B", "C"], 100)

fig = go.Figure()

for cat in ["A", "B", "C"] :
    fig.add_trace(
        go.Scatter(
            x=x_data[categories == cat],
            y=y_data[categories == cat],
            mode="markers",
            name=f"범주 {cat}"
        )
    )

    fig.update_layout(
        title="산점도 그래프 범례 예시",
        legend={
            "title": "범례제목",
            "x":1.01,
            "y":1,
            "xanchor":"left",
            "yanchor":"top",
            "borderwidth": 1,
            "bordercolor": "#000"
        }
    )

fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

x_data = np.arange(50) # 0부터 n-1까지 n개의 정수를 생성한다
y_data = np.random.rand(50) # 0~1 사이의 실수형태의 값을 난수로 50개 생성

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=x_data,
        y=y_data,
        mode="lines",
        line=dict(color="#f00", width=2),
        name="선 그래프",
        showlegend=True
    )
)

fig.update_layout(
    title="선 그래프",
    xaxis_title="시간",
    yaxis_title="값"
)

fig.update_xaxes(
    title_text="시간 (초)",
    title_font={
        "size": 18,
        "color": "purple"
    },
    tickangle=45,
    dtick=5
)

fig.update_yaxes(
    title_text="값 (실수형)",
    title_font={
        "size": 18,
        "color": "#8FD95B"
    },
    tickfont={
        "size": 12
    },
    dtick=0.1
)

fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "A": np.random.rand(50)
})

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=df.index,
        y=df["A"],
        mode="lines+markers",
        name="A 데이터",
        showlegend=True
    )
)

fig.update_layout(
    title={
        "text": "Graph with Scatter",
        "font": {
            "size": 15
        }
    },
    xaxis={"title": "index"},
    yaxis={"title": "A"}
)

fig.show()
```

```python
import plotly.graph_objects as go

categories = ["A", "B", "C", "D"]
values = [23, 17, 35, 29]

fig = go.Figure()

fig.add_trace(
    go.Bar(
        x=categories,
        y=values,
        name="막대 그래프 데이터",
        showlegend=True
    )
)

fig.update_layout(
    title="막대 그래프",
    xaxis_title="범주",
    yaxis_title="값"
)

fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "A": np.random.rand(50),
    "B": np.random.rand(50)
})

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=df.index,
        y=df["A"],
        mode="lines+text",
        name="A",
        showlegend=True
    )
)

fig.add_trace(
    go.Scatter(
        x=df.index,
        y=df["B"],
        mode="lines+markers+text",
        name="B",
        showlegend=True,
        text=df.index,
        textposition="top center"
    )
)

fig.update_layout(
    title="산점도 듀얼 그래프",
    xaxis_title="x축",
    yaxis_title="y축"
)

fig.show()
```
