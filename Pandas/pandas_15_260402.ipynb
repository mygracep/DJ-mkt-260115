```python
import plotly.graph_objects as go

# Figure 생성
fig = go.Figure()

# Figure 안에 그래프 작성 선언
fig.add_trace(
    # 그래프 타입 결정 (* 동시에 여러 그래프를 한 번에 그릴 수 있음)
    go.Scatter(
        # 그래프 내 데이터 삽입
        x=[1, 2, 3],
        y=[1, 2, 3]
    )
)

# 플로틀리 익스프레스 처럼 추가 옵션 설정
# fig.update_layout(), fig.update_xaxes(), fig.update_yaxes()

fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

x_data = np.random.rand(100) # 0~1사이의 실수형태의 난수(랜덤)값 100개
y_data = np.random.rand(100) # 0~1사이의 실수형태의 난수(랜덤)값 100개

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x = x_data,
        y = y_data,
        mode="markers",
        marker=dict(
            size=12,
            color="orange",
            opacity=0.7
        ),
        name="산점도 범례",
        showlegend=True
    )
)

fig.update_layout(
    title="산점도 그래프",
    xaxis_title="x축",
    yaxis_title="y축" 
)

fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

x_data = np.arange(50) # 0부터 n-1까지 n개의 정수를 생성한다
y_data = np.random.rand(50) # 0~1 사이의 실수형태의 값을 난수로 50개 생성

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=x_data,
        y=y_data,
        mode="lines",
        line=dict(color="#f00", width=2),
        name="선 그래프",
        showlegend=True
    )
)

fig.update_layout(
    title="선 그래프",
    xaxis_title="시간",
    yaxis_title="값"
)

fig.show()
```

```python
import plotly.graph_objects as go

categories = ["A", "B", "C", "D"]
values = [23, 17, 35, 29]

fig = go.Figure()

fig.add_trace(
    go.Bar(
        x=categories,
        y=values,
        name="막대 그래프 데이터",
        showlegend=True
    )
)

fig.update_layout(
    title="막대 그래프",
    xaxis_title="범주",
    yaxis_title="값"
)

fig.show()
```
