```python
# Plain Text (특정 파일을 불러오거나, 작성하거나, 편집.수정 작업)
# r(ead) : readlines(), readline(), read()
# w(rite) <- : overwrite
# a(dd | ppend)
```

```python
data_file = open("00_data/text_data_write.txt", "w", encoding="utf-8-sig")
data_file.write("안녕하세요")
data_file.write("David Park 입니다.")
data_file.close()
```

```python
data_file = open("00_data/text_data_write.txt", "w", encoding="utf-8-sig")
data_file.write("안녕하세요\n")
data_file.write("David Park 입니다.\n")
data_file.close()
```

```python
data_file = open("00_data/text_data_write.txt", "a", encoding="utf-8-sig")
data_file.write("해당 파일은 임시 데이터 입니다.")
data_file.close()
```
