```python
# 데이터 분석 : 데이터 찾아와서 수정. 저장
# plain text, csv, xml
# xml = extensible markup language => <product></product>
# html = hyper text markup language => <div></div>

from bs4 import BeautifulSoup

data_file = open("00_data/users.xml", "r", encoding="utf-8-sig")
# print(data_file)
soup = BeautifulSoup(data_file, "xml")
users = soup.select("user")

for user in users :
    print("나이 :", user.select_one("age").text)
    print("이름 :", user.select_one("name").text)
```

```python
import requests
from bs4 import BeautifulSoup

keyword = "나이키"
google_related_keyword = f"https://suggestqueries.google.com/complete/search?output=toolbar&q={keyword}"
response = requests.get(google_related_keyword)
soup = BeautifulSoup(response.content, "xml")

datals = soup.select("suggestion")

for data in datals :
    print(data["data"])
```
