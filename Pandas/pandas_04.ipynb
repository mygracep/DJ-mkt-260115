```python
# json 형태의 파일을 문자열의 형태로 찾아오는 방법
import json

data = '{"id": "01", "language": "Python", "edition": "third", "author": "David Park"}'

jsondata = json.loads(data)
print(jsondata["id"], jsondata["language"], jsondata["edition"])
print(type(jsondata))
```

```python
# json 형태의 파일을 실제 본 쉐입 형태를 유지한 채로 찾아오는 방법
import json

data = {"id": "01", "language": "Python", "edition": "third", "author": "David Park"}
jsondata = json.dumps(data)
print(jsondata, type(jsondata))
```
