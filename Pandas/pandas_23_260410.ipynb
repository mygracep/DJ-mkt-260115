```python
# 날짜에 따른 결제금액을 시계열차트로 정렬해서 시각화!!!
# 해당 데이터의 타입이 문자가 아닌 날짜타입 되어야함!!!
# object => 문자 -> datetime => 날짜타입 변경!!!!

import pandas as pd

df = pd.DataFrame({
    "order": [
        "2020-01-01 07:10:00",
        "2020-01-08 07:20:30",
        "2020-02-20 11:20:00",
        "2020-02-20 04:40:50",
        "2020-02-28 12:10:20",
        "2019-01-10 17:23:50",
        "2019-06-20 22:27:50",
        "2019-06-20 21:15:59",
        "2019-12-10 21:15:59"
    ]
})
```

```python
df.info()
```

```python
import pandas as pd

df["order"] = pd.to_datetime(df["order"], format="%Y-%m-%d %H:%M:%S", errors="raise")
```

### to_datetime() -> 날짜 옵션목록

- %y : 연도를 2글자 표기
- %Y : 연도를 4글자 표기
- %m : 월을 2글자 표기 (01~12)
- %d : 일을 2글자 표기 (01~31)
- %H : 시간을 24시간 기준표기 (00~23)
- %M : 분을 2글자 표기 (00~59)
- %S : 초를 2글자 표기 (00~59)

- errors : 만약 날짜형태로 값을 바꾸기에 불충족한 조건 문자열
- errors=ignore(무시하고 그냥 진행) | raise(에러를 표기하고 나한테 알려줘) | coerce(에러를 NaT 형태로 값을 변환, 진행)
