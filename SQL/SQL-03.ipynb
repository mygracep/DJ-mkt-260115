```python
import pymysql

db = pymysql.connect(
    host="localhost",
    port=3306,
    user="root",
    passwd="",
    db="ecommerce",
    charset="utf8mb4"
)
```

```python
db
```

```python
# 정상적으로 python & mysql 연결
# cursor() 메서드 호출 => Cursor객체 -> mysql DB안에 값을 전송해주는 역할

ecommerce = db.cursor()
```

```python
ecommerce
```

```python
sql = """
    CREATE TABLE product (
        product_code VARCHAR(20) NOT NULL PRIMARY KEY,
        title VARCHAR(200) NOT NULL,
        ori_price INT,
        discount_price INT,
        discount_percent INT,
        delivery VARCHAR(2)
    );
"""
```

```python
sql
```

```python
ecommerce.execute(sql)
```

```python
db.commit()
```

```python
db.close()
```
