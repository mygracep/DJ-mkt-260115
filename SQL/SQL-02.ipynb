# SQL (Structured Query Language) 문법

- DDL (데이터 정의 언어) : Data Definition Language
> CREATE, ALTER, DROP...

- DML (데이터 처리 언어) : Data Manipulation Language
> INSERT, SELECT, GROUP, ORDER...
> 초 / 중 / 고
> 데이터 타입 이해도 : Null, Unknown, 가상쿼리문 & 복합쿼리문 = 서브쿼리
> Create (생성)
> Read (읽기)
> Update (수정)
> Delete (삭제)
 
- DCL (데이터 제어 언어) : Data Control Language
> 계정 -> 변경 | 서브계정 | 팀메이트
> MySQL Cloud

<> : 표준 SQL 문법
!= : MySQL 문법
