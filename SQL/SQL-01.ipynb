*데이터 저장을 위해 알고있어야 하는 중요한 개념

1. 데이터 (Data)
- 관찰, 측정, 조사 등을 통해 수집된 사실, 값, 기록 등의 "원초적인 정보"
- 데이터는 그 자체로는 의미부여가 많이 부족함
- 데이터가 어떠한 가치를 가지려면, 쌓여진 & 축적된 데이터가 하나의 "구조화"가 되어야 비로소 의미있는 가치를 부여
- 월 : 점심 / 화 / 수 / 목 / 금 / 토 / 일 -> 4계절 시간 -> 1년, 3년, 5, 10
- 추측 & 가설

2. 데이터베이스 (Database = DB)
- 여러가지의 다양한 데이터를 "구조적"이고, "체계적"으로 저장 및 관리할 수 있는 저장소
- 2가지 종류
- 1) 관계형 데이터베이스 : RDBMS -> 구조화 되어있는 다양한 데이터들이 서로 관계를 맺고 있는 형태를 의미 (A / B)
> 오랜 역사를 가지고 있음
> 대기업, 글로벌 기업 관계형 데이터베이스 사용
> 안정적, 검증된 데이터 관리 시스템
> 약점 : 서로 다른 데이터가 관계를 맺기 위해서 기본적으로 사용해야는 문법 체계 -> 문법에 의존도가 높음
> 데이터가 우선? VS 데이터를 다루는 언어의 문법 체계가 우선?
> Oracle, MySQL
> SQLD

- 2) 비관계형 데이터베이스 : 구조화 되어있는 데이터들의 관계에 너무 치중하지 않고, 데이터의 입출력 및 수정.편집 등을 원활하게 하기 위하 목적
> CRUD
> Create / Read / Update / Delete
> 스타트업 | 데이터 입출력 굉장히 빈번하게 일어나는 글로벌 기업
> MongoDB
> NoSQL => Not Only SQL
> 문법체계가 유연하다 = 데이터에 포커스가 맞춰짐

> 바이럴 마케팅 -> 컨텐츠 마케팅 -> 퍼포먼스 마케팅 -> 그로스 마케팅 -> 데이터 분석가 -> 데이터 사이언티스트 -> 데이터 엔지니어
> 관계형 데이터베이스
> 관계형 & 비관계형 데이터베이스 이해 & 사용

3. 스키마 (Schema / Scheme)
- 데이터를 체계적으로 구조화한것이 바로 데이터 베이스 -> 데이터 베이스를 생성하기 위해서는 어떤 규칙과 패턴을 가지고 각 데이터들을 체계적으로 구조화했다는 이야기!!!
- 데이터를 데이터베이스화 하기 위한 목적으로 어떤 약속된 구조(틀)을 정의해놓은 것
- 전체 몇개의 컬럼을 기준으로 구조화?
- 각 컬럼안에 입력되어야할 데이터 값이 어떤 타입?
- 데이터의 타입을 세분화해서 메모리를 얼만큼 가용하게 할 것인가?

4. 테이블 (Table)
- 데이터 10만개, 100만개 -> 모두 다 독립적이고, 개별적으로 사용 // 비효율적
- 비슷한 유형 데이터 (1만 / 10만개 -> 소비자, 2만 / 10만개 -> 생산자)
- 데이터를 구조화하기 위해서 데이터 베이스 관리 시스템을 적용해야하는데, 이때 다양한 데이터를 유사한 속성에 따라서 행.열 기반의 테이블로 그룹화할 수 있고 이것을 데이터 베이스 안에 그룹별 저장
- 실질적인 데이터의 저장 단위
- 하나의 데이터 베이스 안에는 복수의 (=여러개의) 테이블이 존재할 수 있음

5. SQL (Structured Query Language)
- 관계형 데이터 베이스를 다루기 위한 문법 체계
- DDL (Data Definition Language) : 생성, 삭제, 교체
- DML (Data Manipulation Language) : 조회, 삽입, 업데이트
- DCL (Data Control Language) : 권한, 박탈, 등급

6. MySQL
- CLI VS GUI
- 명령프롬프트 기반 실행
- GUI : 워크벤치!!!
